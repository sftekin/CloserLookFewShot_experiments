```python
mode = "val"
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
save_path = os.path.join("figures", "{mode}_ens_analysis.png")
for s, focal_div in ens_dict.items():
    stats_arr = np.array(list(focal_div.values()))
    ax.scatter(stats_arr[:, 0], stats_arr[:, 2], stats_arr[:, 1], label=s)

    ax.set_title(f"Ensemble size: {s}, dataset={mode}")
    ax.legend()
    ax.grid()
# plt.savefig(save_path, dpi=200, bbox_inches="tight")
```
